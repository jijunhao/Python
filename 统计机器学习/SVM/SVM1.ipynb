```python
""


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import svm, datasets

xx, yy = np.meshgrid(np.linspace(-3, 3, 500),
                     np.linspace(-3, 3, 500))
#print(xx.ravel())
#print(yy)


#
from sklearn import svm, datasets
data = datasets.make_classification(n_samples=300,n_features=2,n_informative=2,n_redundant=0,n_repeated=0,n_classes=2,
n_clusters_per_class=2,weights=None,flip_y=0.01,class_sep=1.0,hypercube=True,shift=0.0,scale=1.0,shuffle=True,random_state=None)
X = data[0]
Y = data[1]
#np.random.seed(0)
#X = np.random.randn(300, 2)
#Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)    #logical_xor(x1,x2)x1x211


# fit the model
clf = svm.NuSVC(gamma='auto')    #NuSVCSVCgammarbf,polysigmoid
clf.fit(X, Y)
print(clf)


# plot the decision function for each datapoint on the grid
Z0 = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])     #x.ravel()x1
#decison_function

print(Z0)
print(Z0.shape)
Z1 = Z0.reshape(xx.shape)
print(Z1.shape)


#
plt.imshow(Z1, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',
           origin='lower', cmap=plt.cm.PuOr_r)

contours = plt.contour(xx, yy, Z1, levels=[0], linewidths=2,
                       linestyles='dashed')     #

plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap=plt.cm.Paired,
            edgecolors='k')     #s


plt.xticks(())   #x
plt.yticks(())   #y
plt.axis([-3, 3, -3, 3])
plt.show()
```
