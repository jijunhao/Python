```python
import pandas as pd
seeddata = pd.read_csv('seed.txt','\t',header=None)
seeddata
```

```python
from sklearn.linear_model import LogisticRegression
X=seeddata.iloc[:,0:7]
y=seeddata.iloc[:,7]
clf = LogisticRegression(random_state=0).fit(X, y)
print(clf.predict(X[0:2]))

#clf.predict_proba(X[0:2])

print(clf.score(X, y))
```
