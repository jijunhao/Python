# Transform

```python
df[df == 'NR'] = 0    # RAINFALL NR表示未下雨，1表示下雨，将NR变0
data = df.iloc[:, 3:]
data = data.to_numpy(dtype='float')
data
```

```python
data_transform = []   # 将每天离散的数据变成18行连续数据处理,即1号的24点和2号的0点是连续的

for i in range(18):
    data_transform.append([])

for i in range(len(data)):
    data_transform[i%18] += list(data[i])
```

# 标准化

```python
mean_x = np.mean(data_transform,axis=1)
mean_x[9]=0
std_x = np.std(data_transform,axis=1)
std_x[9]=1
```

```python
data_transform_T = (np.array(data_transform).T-mean_x)/std_x
data_transform = data_transform_T.T
```

# 构造训练数据集

```python
x = []
y = []

for i in range(len(data_transform[0])-9):
    x.append(list(data_transform[:,i:i+9].flatten()))
y = np.array(data_transform)[9,9:]
```

```python
train_x = np.c_[np.ones(len(x)),np.array(x)]
train_y = y
```

# 构造测试集

```python
testdf[testdf=='NR']=0
test_x=[]
for name,group in testdf.groupby(0):
    norm = (group.iloc[:,2:].to_numpy(dtype='float').T-mean_x)/std_x
    test_x.append(norm.T.flatten())
test_x = np.c_[np.ones(len(test_x)),np.array(test_x)]
test_x.shape
```

$$y=\sum_{i=1}^{18*9} w_i*x_i+b$$

```python
np.matmul(np.matmul(np.linalg.pinv(np.matmul(train_x.transpose(),train_x)),train_x.transpose()),y)
```

```python
w = np.zeros(len(train_x[0]))
l_rate = 10
repeat = 10000

x_t = train_x.transpose()
s_gra = np.zeros(len(train_x[0]))

for i in range(repeat):
    hypo = np.dot(train_x,w)
    loss = hypo - y
    cost = np.sum(loss**2) / len(train_x)
    cost_a  = np.sqrt(cost)
    gra = np.dot(x_t,loss)
    s_gra += gra**2
    ada = np.sqrt(s_gra)
    w = w - l_rate * gra/ada
    print ('iteration: %d | Cost: %f  ' % ( i,cost_a))
```

```python
test_y=np.dot(test_x,w)
```
