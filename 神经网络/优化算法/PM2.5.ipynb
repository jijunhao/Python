# BGD

```python
 def bgd(x,y,epochs=10000):
    """
    bgd
    """
    w = np.zeros(len(x[0]))
    l_rate = 0.0001    # 学习率
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        grad = np.dot(x.T,error)/len(x)
        w = w - l_rate * grad
        if i%100==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = bgd(x,y)
```

# SGD

```python
def sgd(x,y,epochs=10000):
    """
    sgd
    """
    w = np.zeros(len(x[0]))
    l_rate = 0.0001    # 学习率
    G = np.zeros(len(x[0]))
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        # random
        j = np.random.randint(0, len(x[0]))
        w = w - l_rate * (x[j] * (np.dot(x[j], w) - y[j]))
        if i%100==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = sgd(x,y)
```

```python
w
```

# Mini-batch-gd

```python
def mini(x,y,epochs=100000):
    """
    mini
    """
    w = np.zeros(len(x[0]))
    l_rate = 0.00001    # 学习率
    G = np.zeros(len(x[0]))
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        # 一个个batch算
        j = i%len(x)
        w = w - l_rate * (x[j] * (np.dot(x[j], w) - y[j]))
        if i%1000==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = mini(x,y)
```

```python
w
```

# Adagrad

```python
def adagrad(x,y,epochs=10000):
    """
    adagrad
    """
    w = np.zeros(len(x[0]))
    l_rate = 100    # 学习率
    G = np.zeros(len(x[0]))
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        grad = np.dot(x.T,error)/len(x)
        G += grad**2
        ada = np.sqrt(G+1e-8)
        w = w - l_rate * grad/ada
        if i%100==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = adagrad(x,y)
```

```python
w
```

# RMSprop

```python
def RMSprop(x,y,epochs=10000):
    """
    RMSprop
    """
    w = np.zeros(len(x[0]))
    l_rate = 0.001   # 学习率
    beta = 0.9    # 指数平均权重
    G = np.zeros(len(x[0]))
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        grad = np.dot(x.T,error)/len(x)
        G = beta*G + (1-beta)*grad**2
        ada = np.sqrt(G+1e-8)
        w = w - l_rate * grad/ada
        if i%100==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = RMSprop(x,y)
```

```python
w
```

# Adam

```python
def Adam(x,y,epochs=10000):
    """
    Adam
    """
    w = np.zeros(len(x[0]))
    l_rate = 0.001   # 学习率
    beta1 = 0.9    
    beta2 = 0.99
    G = np.zeros(len(x[0]))
    Mom = np.zeros(len(x[0]))
    
    for i in range(epochs):
        error = np.dot(x,w)-y
        loss = np.dot(error.T,error)/len(x)/2
        grad = np.dot(x.T,error)/len(x)
        
        Mom = beta1*Mom + (1-beta1)*grad
        Mom_ = Mom/(1-beta1**(i+1))
        
        G = beta2*G + (1-beta2)*grad**2
        G_ = G / (1-beta2**(i+1))
        ada = np.sqrt(G_+1e-8)
        
        w = w - l_rate * Mom_/ada
        if i%100==0:
            print ('epoch: %d ======>> Loss: %f ' % (i,loss))
    
    return w
```

```python
w = Adam(x,y)
```
